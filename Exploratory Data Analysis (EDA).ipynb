## Exploratory Data Analysis

```python
# plotting a bar chart for shipmode and it's count

ax = sns.countplot(x = 'shipmode',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for shipmode vs sales

sales_mode = df.groupby(['shipmode'], as_index=False)['sales'].sum().sort_values(by='sales', ascending=False)

sns.barplot(x = 'shipmode',y= 'sales' ,data = sales_mode,hue='shipmode')
```

*From above graphs we can see that most of the buyers choose the standard class mode*

segment 

```python
ax = sns.countplot(data = df, x = 'segment', hue ='shipmode')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
seg_profit = df.groupby(['segment'], as_index=False)['profit'].sum().sort_values(by='profit', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'segment',y= 'profit' ,data =seg_profit,hue='segment')
```

From first we say that most of the consumers,corporates,home office  prefer the standard class shipmode,
we get most profit from the consumers as compared to corporates and home offices
